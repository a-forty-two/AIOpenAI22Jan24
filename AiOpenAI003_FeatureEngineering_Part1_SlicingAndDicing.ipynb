```python
rawdata = {
    "age": [2,3,4,5,5,6,6,42,23,2323,43,435,2],
    "happiness": [12,23,4,5.5,51,6.6,63,42,123,44,41,35,52],
    "dist_from_sea": [2,3,4,5,5,6,6,42,23,2323,43,435,2],
    "dist_from_college": [12,23,4,5.5,51,6.6,63,42,123,44,41,35,52]
}
```

```python
import pandas as pd
# panel dataframes -> rectangular frame of data!
mydata = pd.DataFrame(rawdata)
```

```python
mydata
```

```python
#LOC -> location; iLOC -> INDEX of LOCATION
# syntax-> .loc or .iloc [ row selection, col selection]
# : means all values!
somedata = mydata.iloc[ : , [1,3]]
somedata
```

```python
somedata = mydata.loc[ : , ['happiness','dist_from_college']]
somedata
```

```python
# first 5 rows, first 2 columns
somedata = mydata.iloc[ :5 , :2 ]
somedata
```

```python
# Last 5 rows, last 2 columns
somedata = mydata.iloc[ -5: , -2: ]
# known bug in python libs- if this doesn't work, PLEASE
# don't worry about it!
somedata
```

```python
# from 3rd to 10th row, and
# all columns except first and last
somedata = mydata.iloc[ 3:10 , 1:-1 ]
somedata
```
